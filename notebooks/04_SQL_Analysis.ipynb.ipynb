# SQL Analysis

```python
import sqlite3

import pandas as pd

# Pandas display options
pd.set_option("display.max_columns", None)
```

## Load Clean Dataset

```python
clean_df = pd.read_csv(
    "../data/processed/online_retail_clean.csv",
    dtype={
        "InvoiceNo": str,
        "StockCode": str
    },
    parse_dates=["InvoiceDate"]
)
```

#### Observation

The cleaned dataset created during the previous stage of the project is loaded into pandas before being imported into the SQLite database.

## Create SQLite Database

```python
connection = sqlite3.connect(
    "../data/processed/retail.db"
)
```

#### Observation

A SQLite database is created to store the cleaned dataset and execute SQL queries throughout this notebook.

## Load Data into SQLite

```python
rows_inserted = clean_df.to_sql(
    name="retail",
    con=connection,
    if_exists="replace",
    index=False
)

print(f"Rows inserted into SQLite: {rows_inserted:,}")
```

#### Observation

The cleaned dataset is imported into a SQLite table named **retail**.

This table will be used for all SQL queries presented in this notebook.

## Validation

```python
pd.read_sql(
    """
    SELECT *
    FROM retail
    LIMIT 5;
    """,
    connection
)
```

#### Observation

The first rows confirm that the dataset was successfully loaded into the SQLite database and is ready for analysis.

```python
pd.read_sql(
    """
    SELECT
        COUNT(*) AS TotalRows
    FROM retail;
    """,
    connection
)
```
